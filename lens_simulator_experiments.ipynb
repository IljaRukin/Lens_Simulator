```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

rad2deg = lambda x: x*180/np.pi
deg2rad = lambda x: x*np.pi/180

lamb = 638e-6
z = 500
De = 0.008
#alias = lamb*z/(4*De)
theta0 = np.arctan(lamb/De)
```

```python

G_obj = 10
f = 30
g = 35
lens_pos = 0
b = g*f/(g-f)
theta = 0
print('objekt Size: ',np.round(G_obj,2),' initial angle: ',np.round(theta,2))

G_lens = G_obj+g*np.tan(theta)
plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens])
theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
print('lens diameter: ',np.round(G_lens,2),' angle after lens: ',np.round(theta,2))
G_obj = G_lens+b*np.tan(theta)
plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj])

L = 340
f = 250
g = L - b
lens_pos += L
b = g*f/(g-f)

G_lens = G_obj+g*np.tan(theta)
plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens])
theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
print('lens diameter: ',np.round(G_lens,2),' angle after lens: ',np.round(theta,2))
G_obj = G_lens+b*np.tan(theta)
plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj])

L = 940
f = 50
g = L - b
lens_pos += L
b = g*f/(g-f)

G_lens = G_obj+g*np.tan(theta)
plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens])
theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
print('lens diameter: ',np.round(G_lens,2),' angle after lens: ',np.round(theta,2))
G_obj = G_lens+b*np.tan(theta)
plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj])

plt.show()
```

```python
plt.figure(figsize=(8,4),dpi=200)
plt.axhline(y=0, color='black', linestyle=':')
plt.axis('off')

col = 'orange'
lw = 1
lensadd = 5

for theta in deg2rad(np.linspace(-5,5,5)):
    
    G_obj = 10
    f = 30
    g = 35
    lens_pos = 0
    b = g*f/(g-f)

    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=col,linewidth=lw)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=col,linewidth=lw)
    plt.vlines(lens_pos,-abs(G_lens)-lensadd,abs(G_lens)+lensadd)

    L = 340
    f = 250
    g = L - b
    lens_pos += L
    b = g*f/(g-f)

    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=col,linewidth=lw)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=col,linewidth=lw)
    plt.vlines(lens_pos,-abs(G_lens)-lensadd,abs(G_lens)+lensadd)

    L = 940
    f = 50
    g = L - b
    lens_pos += L
    b = g*f/(g-f)

    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=col,linewidth=lw)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=col,linewidth=lw)
    plt.vlines(lens_pos,-abs(G_lens)-lensadd,abs(G_lens)+lensadd)

plt.show()
```

```python
plt.figure(figsize=(8,4),dpi=200)
plt.axhline(y=0, color='black', linestyle=':')
plt.axis('off')


rayC = 'orange'
rayW = 1

objC = 'green'
imgC = 'red'

LensP = [0,0,0]
LensD = [0,0,0]

for theta in deg2rad(np.linspace(-5,5,5)):
    
    G_obj = 10
    f = 30
    g = 35
    lens_pos = 0
    LensP[0] = lens_pos
    b = g*f/(g-f)
    plt.vlines(-g,0,G_obj,color=objC)
    plt.vlines(b,0,-G_obj*b/g,color=imgC)
    
    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=rayC,linewidth=rayW)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=rayC,linewidth=rayW)
    LensD[0] = max(LensD[0],abs(G_lens))
    
    L = 340
    f = 250
    g = L - b
    lens_pos += L
    LensP[1] = lens_pos
    b = g*f/(g-f)
    plt.vlines(lens_pos-g,0,G_obj,color=objC)
    plt.vlines(lens_pos+b,0,-G_obj*b/g,color=imgC)
    
    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=rayC,linewidth=rayW)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=rayC,linewidth=rayW)
    LensD[1] = max(LensD[1],abs(G_lens))
    
    L = 940
    f = 50
    g = L - b
    lens_pos += L
    LensP[2] = lens_pos
    b = g*f/(g-f)
    plt.vlines(lens_pos-g,0,G_obj,color=objC)
    plt.vlines(lens_pos+b,0,-G_obj*b/g,color=imgC)
    
    G_lens = G_obj+g*np.tan(theta)
    plt.plot([lens_pos-g,lens_pos],[G_obj,G_lens],color=rayC,linewidth=rayW)
    theta = -np.arctan((G_obj*(1+b/g)+g*np.tan(theta))/b)
    G_obj = G_lens+b*np.tan(theta)
    plt.plot([lens_pos,lens_pos+b],[G_lens,G_obj],color=rayC,linewidth=rayW)
    LensD[2] = max(LensD[2],abs(G_lens))

ax = plt.gca()
for p,h in zip(LensP,LensD):
    ellipse = mpl.patches.Ellipse(xy=(p, 0), width=30, height=(h*2)+10)
    ax.add_patch(ellipse)

plt.show()
```

```python
fig = plt.figure(figsize=(8,4),dpi=300)
plt.axhline(y=0, color='black', linestyle=':')
plt.axis('off')

lensNR = 3

f = [0]*lensNR
g = [0]*lensNR
b = [0]*lensNR
G = [0]*lensNR
B = [0]*lensNR
L = [0]*(lensNR-1)

display = [1]*(lensNR+1)

LensP = [0]*lensNR
LensD = [0]*lensNR

#-----

f = [30,250,50]
L = [340,940]

g0 = [34,31]
G0 = [10,-10]
col = ['green','red']
for iii in range(len(g0)):

    g[0] = g0[iii]
    G[0] = G0[iii]
    thetas = deg2rad(np.linspace(-10,10,3))

    display = [1,0,1,1]

    rayC = 'orange'
    rayW = 1

    objC = col[iii]
    imgC = col[iii]

    #-----

    #object/image distances
    for ii in range(lensNR):
        if ii!=0:
            G[ii] = B[ii-1]
            g[ii] = L[ii-1] - b[ii-1]
            LensP[ii] = LensP[ii-1] + L[ii-1]
        b[ii] = g[ii]*f[ii]/(g[ii]-f[ii])
        B[ii] = -G[ii]*b[ii]/g[ii]

    #rays and images
    for theta in thetas:
        for ii in range(lensNR):

            G_lens = G[ii]+g[ii]*np.tan(theta)
            LensD[ii] = max(LensD[ii],abs(G_lens))
            if display[ii]==1:
                plt.vlines(LensP[ii]-g[ii],0,G[ii],color=objC)
                plt.plot([LensP[ii]-g[ii],LensP[ii]],[G[ii],G_lens],color=rayC,linewidth=rayW)

            theta = -np.arctan((G[ii]*(1+b[ii]/g[ii])+g[ii]*np.tan(theta))/b[ii])
            #B[ii] = G_lens+b[ii]*np.tan(theta)
            if display[ii+1]==1:
                plt.vlines(LensP[ii]+b[ii],0,B[ii],color=imgC)
                plt.plot([LensP[ii],LensP[ii]+b[ii]],[G_lens,B[ii]],color=rayC,linewidth=rayW)
            else:
                plt.plot([LensP[ii],LensP[ii+1]],[G_lens,G_lens+L[ii]*np.tan(theta)],color=rayC,linewidth=rayW)

#lenses
ax = plt.gca()
for p,h in zip(LensP,LensD):
    ellipse = mpl.patches.Ellipse(xy=(p, 0), width=30, height=(h*2)*1.3+20, edgecolor='tab:blue', lw=3, facecolor='none')
    ax.add_patch(ellipse)

plt.ylim(-200,200)
plt.show()
```
